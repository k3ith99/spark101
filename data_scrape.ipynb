```python
import pandas as pd
#we scrape the 2022/2023 season first inorder to understand the website layout and how our final dataframe structure will look like
df_2023 = pd.read_html('https://fbref.com/en/comps/11/2022-2023/schedule/2022-2023-Serie-A-Scores-and-Fixtures',index_col=False)[0]
df_2023 = df_2023[df_2023['Wk'].notna()]
df_2023 = df_2023.drop(['Match Report','Notes'],axis=1)
df_2023['HomeGoals'] = [int(i.split('–')[0]) for i in df_2023['Score']]
df_2023['AwayGoals'] = [int(i.split('–')[1]) for i in df_2023['Score']]
df_2023['FinalResult'] = df_2023.apply(lambda row: 'H' if row['HomeGoals'] > row['AwayGoals'] else ('A' if row['HomeGoals'] < row['AwayGoals'] else 'D'), axis=1)
season = f"{int(df_2023['Date'][0][0:4])}-{int(df_2023['Date'][0][0:4])+1}"
df_2023.insert(4,'Season',season)
```

```python
urls = [
    "https://fbref.com/en/comps/11/2013-2014/schedule/2013-2014-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2014-2015/schedule/2014-2015-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2015-2016/schedule/2015-2016-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2016-2017/schedule/2016-2017-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2017-2018/schedule/2017-2018-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2018-2019/schedule/2018-2019-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2019-2020/schedule/2019-2020-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2020-2021/schedule/2020-2021-Serie-A-Scores-and-Fixtures",
    "https://fbref.com/en/comps/11/2021-2022/schedule/2021-2022-Serie-A-Scores-and-Fixtures",
    'https://fbref.com/en/comps/11/2022-2023/schedule/2022-2023-Serie-A-Scores-and-Fixtures'
]
```

```python
#we create a loop that goes through each url in urls and stores each result in a dictionary
dfs = {}
for link in urls:
    df = pd.read_html(link,index_col=False)[0]
    df = df[df['Wk'].notna()]
    df = df.drop(['Match Report','Notes'],axis=1)
    df['HomeGoals'] = [int(i.split('–')[0]) for i in df['Score']]
    df['AwayGoals'] = [int(i.split('–')[1]) for i in df['Score']]
    df['FinalResult'] = df.apply(lambda row: 'H' if row['HomeGoals'] > row['AwayGoals'] else ('A' if row['HomeGoals'] < row['AwayGoals'] else 'D'), axis=1)
    season = f"{int(df['Date'][0][0:4])}-{int(df['Date'][0][0:4])+1}"
    df.insert(4,'Season',season)
    dfs[f'df_{link}'] = df
dfs    
```

```python
#we retrieve the values of each dictionary, combine them and store as a csv to use in our sparktest file
combined_df = pd.concat(dfs.values())
combined_df.to_csv('serieA-matches.csv',header= True, index = True)
combined_df
```
